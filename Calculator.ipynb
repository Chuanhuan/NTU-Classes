
**Normal** **Distribution**

$$
f(x) = \frac{1}{\sqrt{2\pi}\sigma}
        e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^2}
$$

---

**Possion** **Distribution**

$$\begin{aligned}
f(x) &=\frac{e^{-\lambda t}(\lambda t)^x}{x!} \\
\\
&= \frac{e^{-\mu}(\mu)^x}{x!}
\end{aligned}

$$
